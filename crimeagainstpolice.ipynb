```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
import plotly.graph_objects as go 
```

```python
crime =pd.read_csv("13_Police_killed_or_injured_on_duty2001_2010.csv")
crime
```

```python
crime.columns = crime.columns.str.strip()
```

```python
crime.columns
```

```python
crime.isnull().sum()
```

```python
crime_columns=['Police_Injured_By_Criminals', 'Police_Injured_By_Riotous_Mobs',
       'Police_Injured_In_Accidents',
       'Police_Injured_In_Dacoity_OperationsOther_raids',
       'Police_Injured_In_TerroristsExtremists_Operations',
       'Police_Injured_On_Border_Duties', 'Police_Injured_Total_Policemen',
       'Police_Killed_By_Criminals', 'Police_Killed_By_Riotous_Mobs',
       'Police_Killed_In_Accidents',
       'Police_Killed_In_Dacoity_OperationsOther_raids',
       'Police_Killed_In_TerroristsExtremists_Operations',
       'Police_Killed_On_Border_Duties', 'Police_Killed_Total_Policemen']
```

```python
crime_area=crime.groupby('Area_Name',as_index=False)[crime_columns].sum().sort_values(by='Police_Killed_Total_Policemen',ascending=False)
```

```python
crime_year= crime.groupby('Year',as_index=False)[crime_columns].sum().sort_values(by='Police_Killed_Total_Policemen',ascending=False)
```

```python
crime_subgroup =crime.groupby('Sub_Group_Name',as_index=False)[crime_columns].sum().sort_values(by='Police_Killed_Total_Policemen',ascending=False)
```

```python
for crime in crime_columns:
    sorted_crime =crime_area.sort_values(crime,ascending=False)
    fig= px.bar(sorted_crime,x ="Area_Name",y=crime,title = crime.replace("_"," "),
          labels={"Area_Name":"State/ut",crime:"no of cases"},
          hover_data={"Area_Name":True,crime:True})
    
    fig.update_layout(xaxis_title ="State/UT",
                     yaxis_title = crime,
                     yaxis_type ="log",
                     height =800,
                     width =1000)
    fig.show()
    
```

```python
for crime in crime_columns:
    sorted_year =crime_year.sort_values(crime,ascending=False)
    fig = px.bar(sorted_year,y=crime,x ='Year',title =crime.replace("_"," "), 
                color = crime,
                labels={'Year':'Year',crime:crime},
                hover_data={'Year':True,crime:True},
                color_continuous_scale="viridis")
    fig.update_layout(xaxis_title = crime,
                     yaxis_title ="Year",
                     height =600,
                     width =1000)
                    
                     
                    
    fig.show()
```

```python
for crime in crime_columns:
    fig= px.bar(crime_subgroup, x = crime, y ='Sub_Group_Name',title = crime.replace("_"," "),orientation ="h",
               color=crime,
               color_continuous_scale="viridis",
               labels={crime:crime,'Sub_Group_Name':'category'},
               hover_data={crime:True,'Sub_Group_Name':True})
    fig.update_layout(xaxis_title=crime,
                     yaxis_title="category",
                     yaxis=dict(autorange ="reversed"),
                     height =800,
                     width =1000)
    fig.show()
```
